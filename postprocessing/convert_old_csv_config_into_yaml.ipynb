```python
import pandas as pd
import yaml
```

Quick look at the csv generated.

```python
df_network = pd.read_csv('results-normal/1/1.csv')
df_network = df_network.drop(columns=[' wkn', ' M', 'Unnamed: 12'])
df_network
```

```python
df_network.describe()
```

We create a config dictionary that has the right format 

```python
config = {}
config['project name'] = 'adan56_new_open_BF'
config['solver'] = {'Ccfl': 0.9, 'cycles': 100, 'jump': 100, 'convergence tolerance': 1.0}
config['blood'] = {'mu': 4.e-3, 'rho': 1060}
config['network'] = []
```

Small helper dictionary to map the keys from csv to yaml config.

```python
mapper = {k: k[1:] for k in df_network.keys()}
mapper['name'] = 'label'
mapper[' R'] = 'R1'
df_network
```

Add the network of arteries to the config dictionary

```python
for k in df_network.iterrows():
    to_append = {}
    for c, v in k[1].items():
        if c == 'name':
            to_append[mapper[c]] = v
        elif v > 0:
            if c in [' Rp', ' Rd']:
                to_append['R0'] = (k[1][' Rp'] + k[1][' Rd'])/2 # We use constant lumen radius.
            else:
                to_append[mapper[c]] = v
                if c == ' Cc':
                    to_append['outlet'] = 'wk2'
    config['network'].append(to_append)
```

Quick look at it:

```python
config
```

Add some more parameters of the configuration file (inlet flow).

```python
config['network'][0]['inlet'] = 'Q'
config['network'][0]['inlet file'] = 'adan56_inlet.dat'
config['network'][0]['inlet number'] = 1
```

Dump the config dictionary as a yaml file.

```python
with open('config_adan56.yml', 'w') as f:
    yaml.dump(config, f)
```
